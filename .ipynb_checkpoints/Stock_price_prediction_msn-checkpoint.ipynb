```python
# import the data
import pandas as pd
import numpy as np
import math
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense ,LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
msn_df=pd.read_csv('./data/MSN.csv')
```

```python
msn_df.head()
```

```python
msn_df['Date/Time']=pd.to_datetime(msn_df['Date/Time'])
msn_df.sort_values('Date/Time', inplace=True)
```

```python
plt.figure(figsize=(25,15))
plt.title("Close price history")
plt.plot(msn_df['Date/Time'],msn_df['Close'])
plt.xlabel('Date/Time',fontsize=18)
plt.ylabel('Close Price',fontsize=18);
plt.show();
```

```python
data=msn_df.filter(['Close'])
```

```python
dataset=data.values
```

```python
training_data_len=math.ceil(len(dataset)*.8)
```

```python
training_data_len
```

```python
#Scale the data
scaler=MinMaxScaler(feature_range=(0,1))
#fpt
scaled_data=scaler.fit_transform(dataset)
```

```python
scaled_data.shape
```

```python
# create the scaled training dataset
train_data=scaled_data[0 : training_data_len,:]
```

```python
#fpt
x_train=[] #training features
y_train=[] #target 
for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
    if i <=61:
        print(x_train)
        print(y_train)
        print()
```

```python
x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
#reshape the data
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

```python
model=Sequential()
model.add(LSTM(50, return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
#Compile the model
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(x_train,y_train,batch_size=32,epochs=1)
```

```python
#create the testing dataset
#create a new array containing scale values from index 77865 to 97406
test_data=scaled_data[training_data_len - 60: , :]
#create the dataset x_test,y_test
x_test=[]
y_test=dataset[training_data_len:,:]
for i in range(60,len(test_data)):
    x_test.append(test_data[i-60:i,0])
```

```python
#convert the data to numpy array
x_test=np.array(x_test)
```

```python
#reshape the data
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
x_test.shape
```

```python
#get the model predicted the values
predictions=model.predict(x_test)
```

```python
predictions=scaler.inverse_transform(predictions)
```

```python
predictions
```

```python
#get the root mean squared error (RMSE)
rmse=np.sqrt(np.mean(predictions - y_test)**2)
rmse
```

```python
msn_df = msn_df.set_index('Date/Time')
```

```python
#plot the data
train=msn_df[:training_data_len]
valid=msn_df[training_data_len:]
valid['Predictions']=predictions
#visualize the data
plt.figure(figsize=(25,15))
plt.title('Model')
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close price',fontsize=18)
plt.plot(train['Close'])
plt.plot(valid[['Close','Predictions'
                   ]])
plt.legend(['Train','Val','Predictions'],loc='lower right')
plt.show();
```
